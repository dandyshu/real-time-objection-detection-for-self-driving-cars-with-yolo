```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
import cv2
```

```python
label = np.load("../DATA/annotation_list.npy")[()]
```

```python
boxes = [[-1,-1,-1,-1]]
for name in label.keys():
    imgboxes = label[name]
    boxes += imgboxes
boxes = boxes[1:]
```

```python
# analysis all the centers
plt.figure()
for box in boxes:
    cx, cy, w, h = box
    plt.plot(cy, cx, '*')
plt.title("location")

plt.figure()
fig,ax = plt.subplots(1)
for box in boxes:
    cx, cy, w, h = box
    plt.plot(cy, cx, '*')
    rect = patches.Rectangle((0,0),w,h,linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
plt.title('size')
plt.show()
```

```python
wh = np.asarray(boxes)[:,2:4].astype(np.float32)
# define criteria and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(wh,2,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

# Now separate the data, Note the flatten()
A = wh[label.ravel()==0]
B = wh[label.ravel()==1]

# Plot the data
plt.scatter(A[:,0],A[:,1])
plt.scatter(B[:,0],B[:,1],c = 'r')
plt.scatter(center[:,0],center[:,1],s = 80,c = 'y', marker = 's')
plt.xlabel('Height'),plt.ylabel('Weight')
plt.show()
```

```python
plt.figure()
fig,ax = plt.subplots(1)
plt.plot(0, 0, '*')
for box in center:
    w, h = box
    rect = patches.Rectangle((0,0),w,h,linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
plt.title('size')
plt.show()
```
